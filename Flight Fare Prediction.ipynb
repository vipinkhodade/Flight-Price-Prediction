lets separate the Date_of_Journey in day month and year

```python
data['Journey_day']=pd.to_datetime(data['Date_of_Journey'], format='%d/%m/%Y').dt.day
data['Journey_month']=pd.to_datetime(data['Date_of_Journey'], format='%d/%m/%Y').dt.month
data['Journey_year']=pd.to_datetime(data['Date_of_Journey'], format='%d/%m/%Y').dt.year
```

```python
data.head()
```

Now we can drop column Date_of_Journey

```python
data.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
data.head()
```

```python
#Dep_Time is a time when plane leaves  the airport 
#Extracting hour Departure hour
data['Dep_hour']=pd.to_datetime(data.Dep_Time).dt.hour

#Extracting minute
data['Dep_min']=pd.to_datetime(data.Dep_Time).dt.minute
data.head()
```

```python
#No need of Dep_Time, hence dropping Dep_Time column
data.drop('Dep_Time', axis=1, inplace=True)
```

```python
data.head(3)
```

```python
#Arrival_Time is a time when plane arrives the airport 
#Extracting hour Arrival hour
data['Arrival_hour']=pd.to_datetime(data.Arrival_Time).dt.hour

#Extracting hour Arrival minute
data['Arrival_min']=pd.to_datetime(data.Arrival_Time).dt.minute
```

```python
data.head()
```

```python
data.drop('Arrival_Time', axis=1, inplace=True)
```

```python
data.head(2)
```

```python
duration = list(data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(duration[i].split(sep = "m")[0].split()[-1]) # Extracts only minutes from duration
```

```python
data['Duration_hours']=duration_hours
data['Duration_mins']=duration_mins
data.drop('Duration', axis=1, inplace=True)
```

```python
data.head(2)
```

```python
data.Total_Stops.unique()
```

```python
data.Total_Stops.replace({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4}, inplace=True)
```

```python
data['Total_Stops'].fillna(1, inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=0)
```

<h2> Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train, y_train)
y_pred=rfr.predict(x_test)
```

```python
rfr.score(x_train, y_train)
```

```python
rfr.score(x_test, y_test)
```

```python
sns.distplot(y_test-y_pred)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.show()
```

```python
import numpy as np
from sklearn import metrics
```

```python
MAE=metrics.mean_absolute_error(y_test,y_pred)
print('MAE', MAE)
MSE=metrics.mean_squared_error(y_test,y_pred)
print('MSE', MSE)
RMS=np.sqrt(MSE)
print('RMSE', RMS)
r2_score=metrics.r2_score(y_test,y_pred)
print('r2_score' , r2_score)
```
